```python
import yfinance as yf
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Load S&P 500 historical data from 1990 to Feb 2025
def load_sp500_data():
    ticker = "^FTSE"
    start_date = "1990-01-01"
    end_date = "2025-02-01"

    data = yf.download(ticker, start=start_date, end=end_date, auto_adjust=False)

    if data.empty:
        print(f"Error: No data retrieved for {ticker}. Check API availability.")
        return None

    data.reset_index(inplace=True)
    return data

# Plot Close price over time
def plot_sp500_close(data):
    plt.figure(figsize=(14, 7))
    plt.plot(data["Date"], data["Close"], label="FTSE100 Close Price", color="blue")

    plt.gca().xaxis.set_major_locator(mdates.YearLocator(base=5))
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    plt.xticks(rotation=45)
    plt.xlabel("Year")
    plt.ylabel("Close Price")
    plt.title("FTSE100 Close Price")
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()

# Run the script
data = load_sp500_data()
if data is not None:
    plot_sp500_close(data)
```
